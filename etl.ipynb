```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

### Extract CSVs into DataFrames

```python
players_file = "resources/highest_earning_players.csv"
players_df = pd.read_csv(players_file)
players_df.head()
```

```python
sales_file = "resources/vgsales.csv"
sales_df = pd.read_csv(sales_file)
sales_df.head()
```

```python
sales_df = sales_df[["Rank", "Name", "NA_Sales", "EU_Sales", "JP_Sales", "Global_Sales"]]
players_df = players_df[["CurrentHandle", "Game", "TotalUSDPrize", "Genre"]]
sales_df.head()
players_df.head()

sales_df
```

```python
players_df.drop_duplicates("CurrentHandle", inplace=True)
sales_df.drop_duplicates("Name", inplace=True)
sales_df = sales_df.rename(columns={"Name": "Game"})

players_df.set_index("Game", inplace=True)
sales_df.set_index("Game", inplace=True)
```

```python
players_df
```

```python
sales_df
```

```python
engine = create_engine('postgresql://postgres:' + password + '@localhost:5432/videogames')
```

```python
engine.table_names()
```

```python
sales_df.to_sql(name='sales', con=engine, if_exists='append', index=True)
players_df.to_sql(name='players', con=engine, if_exists='append', index=True)
```
